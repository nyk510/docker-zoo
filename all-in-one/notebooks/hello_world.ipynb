```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from atlus import utils, setting
from atlus.visualize import xgb_feature_importance

%matplotlib inline
```

```python
utils.set_default_style()
```

```python
x = np.linspace(-3, 3, 100)
y = np.sin(x) + .3 * np.random.normal(size=x.shape)
```

```python
plt.plot(x, y)
plt.xlabel('えっくすじく')
plt.ylabel('観測値')
plt.title('これがタイトル')
plt.tight_layout()

os.makedirs(setting.FEATURE_DIR, exist_ok=True)
plt.savefig(os.path.join(setting.FEATURE_DIR, 'sample.png'))
```

```python
from sklearn.datasets import load_boston
```

```python
data = load_boston()
```

```python
cols = data['feature_names']
x = data['data']
y = data['target']
```

```python
y = np.where(y > 15., 1, 0)
```

```python
xgb_feature_importance(x, y, cols)
plt.tight_layout()
plt.savefig(os.path.join(setting.FEATURE_DIR, 'bostom__xgb_feature_importance.png'))
```
