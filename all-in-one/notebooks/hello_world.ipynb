```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from atlus import utils, setting

%matplotlib inline
```

```python
utils.set_default_style()
```

```python
x = np.linspace(-3, 3, 100)
y = np.sin(x) + .3 * np.random.normal(size=x.shape)
```

```python
plt.plot(x, y)
plt.xlabel('えっくすじく')
plt.ylabel('観測値')
plt.title('これがタイトル')
plt.tight_layout()

os.makedirs(setting.FEATURE_DIR, exist_ok=True)
plt.savefig(os.path.join(setting.FEATURE_DIR, 'sample.png'), dpi=150)
```
